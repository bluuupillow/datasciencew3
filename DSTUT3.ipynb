```python
import numpy as np

print("1. Create a numpy array of size 2 x 2")
# Method 1: Using np.array()
array_2x2 = np.array([[1, 2], [3, 4]])
print("2x2 array:")
print(array_2x2)
print(f"Shape: {array_2x2.shape}\n")

print("2. Create a random numpy array of size 3 x 4")
random_3x4 = np.random.rand(3, 4)
print("Random 3x4 array:")
print(random_3x4)
print(f"Shape: {random_3x4.shape}\n")

print("3. Create a numpy array of ones and a numpy array of zeroes")
# Ones array
ones_array = np.ones((3, 3))
print("Ones array (3x3):")
print(ones_array)

# Zeros array
zeros_array = np.zeros((2, 4))
print("Zeros array (2x4):")
print(zeros_array)
print()

print("4. Create a random numpy array of size 6x3 and transpose it")
random_6x3 = np.random.rand(6, 3)
print("Original 6x3 array:")
print(random_6x3)
print(f"Original shape: {random_6x3.shape}")

transposed_array = random_6x3.T
print("Transposed array (3x6):")
print(transposed_array)
print(f"Transposed shape: {transposed_array.shape}\n")

print("5. Create two random numpy arrays and compute dot product")
# Define dimensions
n, p, m = 4, 3, 5

# Create random arrays
array1 = np.random.rand(n, p)  # n x p
array2 = np.random.rand(p, m)  # p x m

print(f"Array 1 ({n}x{p}):")
print(array1)
print(f"Array 2 ({p}x{m}):")
print(array2)

# Compute dot product
dot_product = np.dot(array1, array2)
print(f"Dot product result ({n}x{m}):")
print(dot_product)
print(f"Dot product shape: {dot_product.shape}")

# Alternative method using @ operator
dot_product_alt = array1 @ array2
print("\nUsing @ operator (same result):")
print(dot_product_alt)
```
